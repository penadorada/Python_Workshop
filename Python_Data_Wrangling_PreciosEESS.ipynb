```python
#Seleccionamos las Petroleras con más de 12 gasolineras
oil = ['REPSOL','CEPSA','GALP','BP','SHELL','CAMPSA','PETRONOR','AVIA','CARREFOUR','DISA','BALLEMOIL','SARAS','MEROIL','EROSKI',
       'BONAREA','AGLA','PETROCAT','PETROPRIX','ESCLATOIL','VALCARCE','ALCAMPO','IBERDOEX','GASEXPRESS','TAMOIL','SIMPLY','PETROMIRALLES',
       'PCAN','E.LECLERC','BEROIL']
est_marcas = df[df.Petrolera.isin(oil)]
```

```python
#Se pierden datos de 2638 Gasolineras. Independientes y estaciones de servicio que no muestran que marca distribuyen
est_marcas.shape
```

```python
#Petroleras con Gasóleo más económico de media: 
est_marcas.groupby('Petrolera',as_index=False).Precio_Gasoleo_A.agg(['max','min','mean']).sort_values(by='mean',ascending=True)
```

```python
# Averiguamos donde se localizan los tres distribuidores de Gasóleo A más económicos : 
#Cataluña,Aragón e Islas Canarias (Fiscal)

baratas =est_marcas[(est_marcas.Petrolera=='DISA') | (est_marcas.Petrolera=='PCAN') | (est_marcas.Petrolera=='BONAREA')]
baratas.groupby(['Petrolera','Provincia'],as_index=False).Precio_Gasoleo_A.mean()
```

```python
#Localización Distribuidores con Gasolina 95 más económica.

Gas =est_marcas[(est_marcas.Petrolera=='DISA') | (est_marcas.Petrolera=='PCAN') | (est_marcas.Petrolera=='BONAREA') |
               (est_marcas.Petrolera=='E.LECLERC')]
Gas.groupby(['Petrolera','Provincia'],as_index=False).Precio_Gasolina_95.mean().sort_values(by='Precio_Gasolina_95',ascending= True)
```
